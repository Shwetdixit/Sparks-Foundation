# Data Visualization

```python
# Heat Map for Correlation between Variables
fig,axes = plt.subplots(1,1,figsize=(12,7))
datacorr = data_df.corr()
sns.heatmap(datacorr, annot=True, cmap='RdYlGn')
plt.title('Correlation between variables')
plt.show()
```

# EXPLORATORY DATA ANALYSIS

```python
# Pair Plotting
sns.pairplot(data_df)
```

# A- Ship Mode wise Profit

```python
# Finding out which segment has the highest and lowest profit
p = data_df.groupby(by ='Ship Mode')[['Profit']].sum()
r = p.sort_values(by = ['Profit'],ascending = True)
r.plot(kind = 'bar',figsize =(7,5),color='yellow')
plt.title('Shipping Modes wise Profit')
plt.show()
```

OBSERVATION : From the above plot we can observe that 'Standard Class' has the highest profit where as 'Same Day' has the lowest profit.

# B- Segment Wise Profit

```python
# Finding out which segment has the highest and lowest profit
p = data_df.groupby(by ='Segment')[['Profit']].sum()
r = p.sort_values(by = ['Profit'],ascending = True)
r.plot(kind = 'bar',figsize =(7,5),color='green')
plt.title('Segment wise Profit')
plt.show()
```

OBSERVATION : From the above plot, we can observ that 'Consumer' has the highest profit where as 'Home Office' has the lowest profit. So to increase profit we must focus on 'Home Office and Corporate Segment'.

# C- City wise Profit

```python
data_df.groupby(by ='City')['Profit'].sum().sort_values(ascending = False)[521:].plot(kind = 'bar',color='purple')
plt.show()
```

```python
data_df.groupby(by ='City')['Profit'].sum().sort_values(ascending = False)[:11].plot(kind = 'bar',color='purple')
plt.show()
```

# D- States wise Profit

```python
# State vs Profit
plt.figure(figsize =(20,12))
data_df.groupby(by ='State')['Profit'].sum().sort_values(ascending = True).plot(kind = 'bar')
plt.show()
```

# E- Region wise Segment's Sales Comparison

```python
## let's compare state vs profit
plt.figure(figsize=(10, 7))
sns.barplot(x="Region", y="Sales", hue="Segment", data=data_df,palette='rainbow')
plt.title('Region wise Segments Profit')
plt.show()
```

# F- Region wise Profit

```python
plt.figure(figsize =(8,6))
data_df.groupby(by ='Region')['Profit'].sum().plot(kind = 'pie',autopct='%1.1f%%',explode=(0.1, 0, 0,0),colors = ['lightcoral', 'red', 'yellowgreen', 'blue'])
plt.show()
```

OBSERVATION : From the above pie chart, we can state that the 'Central Region' has the lowest profit whereas the 'West Region' has the highest profit.

So in order to increase profit we must focus on regions with lesser profits like Central and South region.

# G- Category wise Sales Comparison

```python
plt.figure(figsize=(10, 7))
sns.barplot(x="Region", y="Sales", hue="Category", data=data_df,palette="YlOrBr")
plt.title("Region wise Category's Profit")
plt.show()
```

# H- Sub-Category wise Sales and Profit

```python
subcategory_grpby = data_df.groupby(by ='Sub-Category')[['Sales','Profit']].sum()
sort_profit2 = subcategory_grpby.sort_values(by = ['Profit'],ascending = True)
sort_profit2.plot(kind = 'bar',figsize =(13,7))
plt.show()
```

# I- Discount Vs Profi

```python
plt.figure(figsize=(12,5))
sns.lineplot(x='Discount',y='Profit',data=data_df)
plt.title("Discount and Profit")
```

OBSERVATION : There is a negative correlation between Discount and Profit. As discount increases Profit decreases.
